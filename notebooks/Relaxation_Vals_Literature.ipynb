```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
CSF = pd.read_excel('T1_CSF.xlsx', sheet_name='CSF')
WM = pd.read_excel('T1_CSF.xlsx', sheet_name='WM')
```

```python
fig = plt.figure(figsize=(8,3))
fig.add_subplot(1,2,1)
# plt.plot(CSF.Field, CSF.T1, 'ok', markersize=7)
plt.semilogx(CSF.Field, CSF.T1, 'ok', markersize=5)
plt.xlabel('Field Strength [T]')
plt.ylabel('T1 [ms]')
plt.title('CSF')
plt.ylim([0,6000])
plt.grid('both')

fig.add_subplot(1,2,2)
# plt.plot(WM.Field, WM.T1, 'ok', markersize=7)
plt.semilogx(WM.Field, WM.T1, 'ok', markersize=5)
plt.xlabel('Field Strength [T]')
plt.ylabel('T1 [ms]')
plt.title('White Matter')
plt.ylim([0,2000])
plt.tight_layout()
plt.grid('both')

plt.savefig('WM_CSF_T1_B0.png', dpi=300)
```
